## Data Wrangling


### General Properties

```python
# Load your data and print out a few lines. Perform operations to inspect data
#   types and look for instances of missing or possibly errant data.
#load data 
df = pd.read_csv(r"C:\Users\ABO_HAGAG\tmdb-movies.csv")
#print first rows in file 
df.head()

```



### Data Cleaning (Delete unused columns, delete duplicates, and modify the structure)
At this stage i will:
1- Delete unused columns (cast, homepage, tagline, keywords, etc)
2-Delete redundancy
3-Delete null or 0 values

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.
#Delete unused columns (cast, homepage, tagline, keywords, etc)
df=df.drop(['keywords', 'imdb_id', 'tagline', 'cast', 'overview', 'homepage'], axis=1)
#Delete redundancy
sum(df.duplicated())
df.drop_duplicates()
#Delete null or 0 values
df.dropna(inplace=True)
#replacing 0 values with NAN
df.replace(0, np.NAN)
#handeling missing values
df['budget'] = df['budget'].replace(np.NAN,df['budget'].mean())
df['revenue'] = df['revenue'].replace(np.NAN,df['revenue'].mean())
df['runtime'] = df['runtime'].replace(np.NAN,df['runtime'].mean())
df.head(4)
```

## Exploratory Data Analysis


### Research Question 1 (What is the relation between number of movies and years?)

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and finding

#knowing numbers of moves per year 
years_count = df["release_year"].value_counts().sort_index()

# showing number of movies per year 
years_count.plot(kind='barh',title='moves number & year ', figsize=(10,25))
# x-label name
plt.xlabel(' Movies number')
# y-label name
plt.ylabel('year')

 
```

in above plot, we can see that the number of movies increases and the year that has hightiest number of movies is 2014

### Research Question 2  (how long is the average runtime of movies?)

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.
#getting the movies avarge time 
plt.figure(figsize=(10,5))
#setting title to the plot 
plt.title('the movies avarge runtimes ')
#y-label name 
plt.ylabel('movies numbers')
#x-lable name 
plt.xlabel('runtime')
plt.hist(df['runtime'])
df["runtime"].describe()
```
